```python
data_dict = maya.to_dict(orient="index")
```

```python
data_dict["GT0001"]
```

```python
for sim in data_dict.keys():
    data_dict[sim]["a1"] = [data_dict[sim].pop("a1x"), data_dict[sim].pop("a1y"), data_dict[sim].pop("a1z")]
    data_dict[sim]["a2"] = [data_dict[sim].pop("a2x"), data_dict[sim].pop("a2y"), data_dict[sim].pop("a2z")]
```

```python
vectors = {
    "a1": ["a1x", "a1y", "a1z"],
    "a2": ["a2x", "a2y", "a2z"],
}

for values in data_dict.values():
    for spin_label, components in vectors.items():
        values[spin_label] = [values.pop(c) for c in components]
```
